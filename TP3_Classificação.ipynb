```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier


```

```python
url = 'https://raw.githubusercontent.com/cassiusf/datasets/main/customers_data.csv'

data = pd.read_csv(url)
data.head()
```

```python
# 1. Verificar e eliminar dados ausentes
data = data.dropna()

# 2. Recodificar variável alvo (1 = Horeca, 0 = Retail)
data['Channel'] = data['Channel'].replace({2: 0})

# Separar features e target
X = data.drop(['Channel'], axis=1)
y = data['Channel']
```

```python

# 3. Dividir em treino e teste (75% treino / 25% teste)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Tamanho do treino:", X_train.shape[0])
print("Tamanho do teste:", X_test.shape[0])
print("Proporção de classes no treino:")
print(y_train.value_counts(normalize=True))
```

```python
# Criar e treinar o modelo Random Forest (opções padrão)
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Prever no conjunto de teste
y_pred = rf.predict(X_test)

# Avaliar desempenho
print("Acurácia no teste:", accuracy_score(y_test, y_pred))
print("\nMatriz de Confusão:\n", confusion_matrix(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred))
```

```python
# Criar e treinar o modelo XGBoost (parâmetros padrão)
xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)

# Prever no conjunto de teste
y_pred_xgb = xgb.predict(X_test)

# Avaliar desempenho
print("Acurácia no teste:", accuracy_score(y_test, y_pred_xgb))
print("\nMatriz de Confusão:\n", confusion_matrix(y_test, y_pred_xgb))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred_xgb))
```

Questão: Qual modelo você utilizaria, a partir dos resultados da execução dos dois modelos? Justifique sua resposta.


Eu escolheria o Random Forest porque, nos parâmetros padrão, ele apresentou maior acurácia geral e desempenho mais equilibrado entre as classes. O XGBoost tem potencial para superar o RF após tuning, mas como a análise pediu execução com configurações padrão, o RF se mostrou a melhor opção nesse caso.
